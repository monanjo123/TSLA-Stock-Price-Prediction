```python
import time
import os
import random
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import statistics
import numpy as np
import scipy
import seaborn as sns
import copy
import yfinance as yf
from pandas import concat, DataFrame
from sklearn.linear_model import RANSACRegressor, LinearRegression, TheilSenRegressor
from sklearn.metrics import explained_variance_score, mean_absolute_error, mean_absolute_error, median_absolute_error, r2_score
from sklearn.svm import SVR
from sklearn.linear_model import Ridge, Lasso, ElasticNet, BayesianRidge
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.pipeline import Pipeline
from pandas_datareader import data as pdr
from sklearn.neural_network import MLPRegressor
```

# Get Dataset

```python
if(os.path.exists('TSLA.csv')):
    df_full = pd.read_csv('TSLA.csv')
else:
    yf.pdr_override()
    df_full = pdr.get_data_yahoo("TSLA", start="2010-01-01").reset_index()
    df_full.to_csv('TSLA.csv',index=False)

df_full.head()
```

# Explore Data

```python
df_full['Adj Close'].plot(label='TSLA', figsize=(16,8), title='Adjusted Closing Price', grid=True)
plt.show()
```

```python
plot_count = 1
for feature in list(df_full.columns)[1:]:
    fig, ax = plt.subplots()
    df_full[feature].plot(label='TSLA', figsize=(15,10.5),grid=True, title=feature, subplots=True, ax=ax)
    plot_count += 1
plt.show()
```

```python
window_size = 32 # Allow us to look at 32 days into the past
# Prepare the data so that we have 32 day windows and predict what the next day should be

# Get indices of access for the data
num_samples = len(df_full) - window_size
indices = np.arange(num_samples).astype(np.int)[:,None] + np.arange(window_size + 1).astype(np.int)
data = df_full['Adj Close'].values[indices] # Create the 2D matrix of training samples
X = data[:,:-1] # Each row represents 32 days in the past
y = data[:,-1] # Each output value represents the 33rd day
```

```python
# Preprocess Data
num_samples = len(df_full)
train_index = 0.8

remove_features = ['Open', 'High', 'Low', 'Close', 'Volume']

split = int(num_samples * train_index)
training_data = df_full[:split]
test_data = df_full[split:]
print(len(training_data), len(test_data))
x_train = X[:split]
y_train = y[:split]
x_test = X[split:]
y_test = y[split:]
print(len(x_train), len(y_train), len(x_test), len(y_test))
```

# Using Multiple Linear Regression

```python
# Create Linear Regression
model = LinearRegression()

# train the model using the training sets
model.fit(x_train, y_train)

y_pred_line = model.predict(x_test)

df_line = test_data.copy()

df_line.drop(remove_features, axis=1, inplace=True)
df_line = df_line.iloc[window_size:]

# print(df_line['Adj Close'][:5])
df_line['Adj Close Prediction'] = y_pred_line

df_line.set_index('Date', inplace=True)
df_line.plot(figsize=(16,8))
plt.show()
```

```python
# Create Ridge Regression Model
model = Ridge()

# train the model using the training sets
model.fit(x_train, y_train)

y_pred_ridge = model.predict(x_test)

df_ridge = test_data.copy()

df_ridge.drop(remove_features, axis=1, inplace=True)
df_ridge = df_ridge.iloc[window_size:]


df_ridge['Adj Close Prediction'] = y_pred_ridge

df_ridge.set_index('Date', inplace=True)
df_ridge.plot(figsize=(16,8))
plt.show()
```

```python
# Create ElasticNet Model
model = ElasticNet()

# train the model using the training sets
model.fit(x_train, y_train)

y_pred_elastic = model.predict(x_test)

df_elastic = test_data.copy()

df_elastic.drop(remove_features, axis=1, inplace=True)
df_elastic = df_elastic.iloc[window_size:]


df_elastic['Adj Close Prediction'] = y_pred_elastic

df_elastic.set_index('Date', inplace=True)
df_elastic.plot(figsize=(16,8))
plt.show()
```

```python
# Create Gradient Boosting Regressor Model
model = GradientBoostingRegressor()

# train the model using the training sets
model.fit(x_train, y_train)

y_pred_gbr = model.predict(x_test)

df_gbr = test_data.copy()

df_gbr.drop(remove_features, axis=1, inplace=True)
df_gbr = df_gbr.iloc[window_size:]

# print(df_line['Adj Close'][:5])
df_gbr['Adj Close Prediction'] = y_pred_gbr

df_gbr.set_index('Date', inplace=True)
df_gbr.plot(figsize=(16,8))
plt.show()
```

# MLPRegressor

```python
# Create MLP Regressor Model
model = MLPRegressor(hidden_layer_sizes=(100, 200, 100), activation='relu', 
                         solver='lbfgs', alpha=0.010, learning_rate_init = 0.001, shuffle=False)

model.fit(x_train, y_train)

y_pred_mlp = model.predict(x_test)

df_mlp = test_data.copy()

df_mlp.drop(remove_features, axis=1, inplace=True)
df_mlp = df_mlp.iloc[window_size:]

# print(df_line['Adj Close'][:5])
df_mlp['Adj Close Prediction'] = y_pred_mlp

df_mlp.set_index('Date', inplace=True)
df_mlp.plot(figsize=(16,8))
plt.show()
```

# Combine all on single graph

```python
df_all = test_data.copy()
df_all.drop(remove_features, axis=1, inplace=True)
df_all = df_all.iloc[window_size:]
df_all['Adj Close Gradient Boosting'] = y_pred_gbr
df_all['Adj Close MLP'] = y_pred_mlp
df_all['Adj Close Elastic'] = y_pred_elastic
df_all['Adj Close Ridge'] = y_pred_ridge
df_all['Adj Close Linear'] = y_pred_line
df_all.set_index('Date', inplace=True)
df_all.plot(figsize=(16,8))
plt.show()
```
